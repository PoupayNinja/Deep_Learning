### I.2.b Générer un mot 

```python
def etat_suivant(etat, matrice_transitions):
    transition_ligne = matrice_transitions[etat-1]  # extraire la ligne qui correspond à l'état courant
    cum_probs = np.cumsum(transition_ligne)  # calculer la somme cumulée de probabilités
    rand_num = np.random.random()  # générer un nombre aléatoire entre zéro et un
    etat_suivant = np.argmax(cum_probs >= rand_num) + 1  # choisir l'état suivant
    return etat_suivant
```

```python
def afficher_fonction_repartition(etat, matrice_transitions):
    transition_ligne = matrice_transitions[etat-1, :]
    cum_probs = np.cumsum(transition_ligne)
    etats = range(1, len(transition_ligne) + 1) 
    plt.step(etats, cum_probs)
    plt.title('fonction de repartition $F_{X}(x)$')
    plt.show()
```

```python
def genere_state_seq(matrice_transitions): # générer une séquence d'états
    etat_courant = 1
    seq_etats = [1]
    while etat_courant != len(matrice_transitions): # tant qu'on n'a pas atteint l'état final, on continue
        etat_courant = etat_suivant(etat_courant, matrice_transitions)
        seq_etats.append(etat_courant)
    return seq_etats
```

```python
def display_seq(seq_etats, dic): # afficher une séquence d'états
    return ''.join([dic[str(etat)] for etat in seq_etats])
```

```python
def calc_vraisemblance(phrase, dic_inv, matrice_transitions):
  sequence = []
  mots = phrase.split()
  for mot in mots:
      sequence += ['-{}+'.format(mot)]
  sequence[-1] += '.'
  merged_words = ''.join([seq for seq in sequence])
  symboles = [int(dic_final_inv[letter]) - 1 for letter in merged_words]

  vraisemblance = 1
  for i in range(len(symboles) - 1):
      transition = (symboles[i], symboles[i+1])
      vraisemblance *= matrice_transitions[transition]
    
  return vraisemblance
```

#### II.2.2 Ecrire une fonction etat_suivant qui génère un état qt+1(à t+1) à partir de l’état courant qt (à t) à l’aide de la matrice de transitions et de la fonction de répartition cumsum. Afficher la fonction de répartition pour une ligne de la matrice de transition et expliquer son rôle pour la génération de l’état à t+1.

```python
def etat_suivant(etat, matrice_transitions):
    transition_ligne = matrice_transitions[etat-1]  # extraire la ligne qui correspond à l'état courant
    cum_probs = np.cumsum(transition_ligne)  # calculer la somme cumulée de probabilités
    rand_num = np.random.random()  # générer un nombre aléatoire entre zéro et un
    etat_suivant = np.argmax(cum_probs >= rand_num) + 1  # choisir l'état suivant
    return etat_suivant
```

```python
def afficher_fonction_repartition(etat, matrice_transitions):
    transition_ligne = matrice_transitions[etat-1, :]
    print(f"la ligne de transition est: {transition_ligne}")
    cum_probs = np.cumsum(transition_ligne)
    etats = range(1, len(transition_ligne) + 1) 
    plt.step(etats, cum_probs)
    plt.title('fonction de repartition $F_{X}(x)$')
    plt.show()
```

#### II.2.3 Générer une séquence d’observations suivant le modèle de Markov Caché du chiffre 0. On commencera par générer une séquence d’états suivant ce modèle à l’aide de la fonction etat_suivant. Puis on générera la séquence d’observations par le même procédé.

```python
def genere_state_seq(matrice_transitions): # générer une séquence d'états
    etat_courant = 1
    seq_etats = [1]
    while etat_courant != len(matrice_transitions): # tant qu'on n'a pas atteint l'état final, on continue
        etat_courant = etat_suivant(etat_courant, matrice_transitions)
        seq_etats.append(etat_courant)
    return seq_etats
```

```python
def genere_ob(etat, matrice_prob_obs): # générer une observation
  symboles = matrice_prob_obs[:, etat-1]  
  cum_probs = np.cumsum(symboles)  
  rand_num = np.random.random()  
  ob = np.argmax(cum_probs >= rand_num) + 1  
  return ob
```

```python
def genere_ob_seq(matrice_transitions, matrice_prob_obs): # générer une séquence d'observations
    etat_courant = 1
    seq_etats = [1]
    seq_obs = []
    ob = genere_ob(etat_courant, matrice_prob_obs)
    seq_obs.append(ob)
    while etat_courant != len(matrice_transitions):      
      etat_courant = etat_suivant(etat_courant, matrice_transitions)
      seq_etats.append(etat_courant)     
      ob = genere_ob(etat_courant, matrice_prob_obs)
      seq_obs.append(ob)   
    return seq_etats, seq_obs
```

#### II.3.1 Calculer la vraisemblance de ces séquences suivant chacun des modèles (0, 1 et 7) par l’algorithme de Viterbi (on pourra implémenter la version logarithmique de cet algorithme). Pour cela les matrices A, B et p seront converties en logarithmes (utiliser np.log). 

```python
# Définir la fonction de l'algorithem Viterbi
def viterbi(obs, states, start_p, trans_p, emit_p):
    """
    Retourne la séquence d'états la plus probable correspondant aux observations données en utilisant
    l'algorithme de Viterbi pour un Modèle de Markov Caché.

    obs : une séquence d'observations
    états : une séquence d'états possibles
    start_p : un dictionnaire de probabilités de départ pour chaque état
    trans_p : une matrice de probabilités de transition entre les états
    emit_p : une matrice de probabilités d'émission de chaque état pour chaque observation.
    """
    V = [{}]
    path = {}

    # Initialisation t = 0
    for state in states:
        V[0][state] = emit_p[obs[0]][state] * start_p[state]   
        path[state] = [state]

    # t > 0
    for t in range(1, len(obs)):
        V.append({})
        new_path = {}

        for state in states:
            # Trouver la probabilité maximale et l'état correspondant du pas de temps précédent
            (prob, prev_state) = max((V[t-1][prev_state] + np.log(trans_p[prev_state][state] + 1e-10)  + np.log(emit_p[obs[t]][state] + 1e-10), prev_state) for prev_state in states)

            # Mettre à jour les informations de probabilité et de chemin pour cet état
            V[t][state] = prob
            new_path[state] = path[prev_state] + [state]

        # Mettre à jour le chemin avec les nouvelles informations sur le chemin
        path = new_path

    # Trouver l'état final avec la plus grande probabilité
    (prob, state) = max((V[len(obs)-1][state], state) for state in states)
    prob = np.exp(prob)

    # Retourner en arrière pour trouver la séquence d'états la plus probable
    state_seq = [state]
    for t in range(len(obs)-1, 0, -1):
        state = path[state][t]
        state_seq.insert(0, state)

    # Retourner la séquence d'états la plus probable et sa probabilité
    return pd.DataFrame.from_dict(V), state_seq, prob
```

```python
seq_file_names = ['SeqTest0.txt', 'SeqTest1.txt', 'SeqTest7.txt' ]

def compute_likehoods(seq_file_names): 
  inds =[0, 1, 7]

  # charger les séquences de text
  A_file_names = ['A0.txt', 'A1.txt', 'A7.txt']
  B_file_names = ['B0.txt', 'B1.txt', 'B7.txt']
  pi_file_names = ['vect_pi0.txt', 'vect_pi1.txt', 'vect_pi7.txt']
  
  liste_likehoods = [] 
  for k in range(len(seq_file_names)) :
    A = np.loadtxt(os.path.join(path_to_folder + A_file_names[k]))
    B = np.loadtxt(os.path.join(path_to_folder + B_file_names[k]))
    pi = np.loadtxt(os.path.join(path_to_folder + pi_file_names[k]))
    
    test_sequence = np.loadtxt(os.path.join(path_to_folder + seq_file_names[k])) 

    # Faire une boucle sur chaque séquence de test et calculer sa vraisemblance 
    test_sequence = test_sequence.astype('int')
    for j in range(10):
      sequence = [i-1 for i in test_sequence[j] ]
      _, _, likehood = viterbi(sequence, range(len(A)), pi, A, B)
      print(f"Likelihood for the sequence {j} for Number {inds[k]} is : {likehood} ") 
      liste_likehoods.append(likehood)
  return liste_likehoods
```

#### II.3.2 Donner le résultat de la classification des images de test en considérant un problème à trois classes : 0, 1 et 7. 

```python
seq_file_names = ['SeqTest0.txt', 'SeqTest1.txt', 'SeqTest7.txt' ]

def make_classification(seq_file_names): 
  inds =[0, 1, 7]

  result = pd.DataFrame(columns = ["Modèle0","Modèle1", "Modèle7"])

  test_sequences = []
  for k in range(len (seq_file_names)) : 
    test_sequence = np.loadtxt(os.path.join(path_to_folder + seq_file_names[k])) 
    test_sequences.append(test_sequence)

  A0 = np.loadtxt(os.path.join(path_to_folder + 'A0.txt'))
  A1 = np.loadtxt(os.path.join(path_to_folder + 'A1.txt'))
  A7 = np.loadtxt(os.path.join(path_to_folder + 'A7.txt'))

  B0 = np.loadtxt(os.path.join(path_to_folder + 'B0.txt'))
  B1 = np.loadtxt(os.path.join(path_to_folder + 'B1.txt'))
  B7 = np.loadtxt(os.path.join(path_to_folder +'B7.txt'))

  pi0 = np.loadtxt(os.path.join(path_to_folder + 'vect_pi0.txt'))
  pi1 = np.loadtxt(os.path.join(path_to_folder + 'vect_pi1.txt'))
  pi7 = np.loadtxt(os.path.join(path_to_folder + 'vect_pi7.txt'))
  
  for test_sequence in test_sequences : 
    test_sequence = test_sequence.astype('int')
    for j in range(10):
      liste_likehoods = [] 
      sequence = [i-1 for i in test_sequence[j] ]
      _, _, likehood0 = viterbi(sequence, range(len(A0)), pi0, A0, B0)
      _, _, likehood1 = viterbi(sequence, range(len(A0)), pi1, A1, B1)
      _, _, likehood7 = viterbi(sequence, range(len(A0)), pi7, A7, B7)

      liste_likehoods.append(likehood0)
      liste_likehoods.append(likehood1)
      liste_likehoods.append(likehood7)

      result.loc[len(result)] = liste_likehoods

  return result

probas = make_classification(seq_file_names).to_numpy()
output = pd.DataFrame(columns = ["Predictions"])

# Classifier la séquence de test sur la base du modèle le plus probable

max_likelihood_index = np.argmax(probas, axis=1)
max_likelihood_index = [7 if i == 2 else i for i in max_likelihood_index]
output = pd.DataFrame(max_likelihood_index, columns = ["Predictions"])
display(output)
```
